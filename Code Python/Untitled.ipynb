```python
import numpy as np
import matplotlib.pyplot as plt

verbose = False
```

```python
def create_ElemList(ElemList, numberElem, NodeList):
    
    newElemList = []
    
    for elem in ElemList:
        i = elem[0]
        j = elem[1]
        propriety = elem[2]
        
        coord1 = NodeList[i]
        coord2 = NodeList[j]
        
        current = i
        if propriety != 2:
            
            dx = (coord2[0] - coord1[0]) / numberElem
            dy = (coord2[1] - coord1[1]) / numberElem
            dz = (coord2[2] - coord1[2]) / numberElem
            
            for i in range(numberElem-1):
                newElem = len(NodeList)
                
                newElemList.append([current, newElem, propriety])
                
                coord = NodeList[current]
                NodeList.append([coord[0] + dx, coord[1] + dy, coord[2] + dz])
                
                current = newElem
            
            newElemList.append([current, j, propriety])
            
        else:
            newElemList.append(elem)
            
    return newElemList

def create_dofList(NodeList):
    
    dof = 0
    dofList = []
    
    for i in range(len(NodeList)):
        dofs = []
        for i in range(6):
            dofs.append(dof)
            
            dof += 1
        dofList.append(dofs)
    
    return dofList

def create_locel(ElemList, DofList):
    
    locel = []
    
    for elem in ElemList:
        
        dof1 = DofList[elem[0]]
        dof2 = DofList[elem[1]]
        
        locel.append(dof1+dof2)
        
    return locel
```
